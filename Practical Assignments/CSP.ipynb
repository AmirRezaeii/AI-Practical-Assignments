### Backtracking Algorithm
Testing your algorithm on a sample sudoku is enough, but your code will be manually tested on multiple ones later

```python
def is_empty(sudoku):
    for row in range(9):
        for col in range(9):
            if sudoku[row, col] == 0:
                return (row, col)
    return None
```

```python
def is_placement_valid(grid, num, position):
    row, col = position
    if num in grid[row, :]:
        return False
    if num in grid[:, col]:
        return False
    box_x = (row // 3) * 3
    box_y = (col // 3) * 3
    if num in grid[box_x:box_x + 3, box_y:box_y + 3]:
        return False
    return True
```

```python
def backtrack(sudoku):
    empty = is_empty(sudoku)
    if not empty:
        return sudoku
    row, col = empty
    for num in range(1, 10):
        if is_placement_valid(sudoku, num, (row, col)):
            sudoku[row, col] = num
            if backtrack(sudoku) is not None:
                return sudoku
            sudoku[row, col] = 0
    return None
```

```python
def compare(sudoku, solution):
    difference = sudoku - solution
    return np.all(difference == 0)

def data_to_sudoku(entry):
    parsed_numbers = [int(char) for char in entry]
    while len(parsed_numbers) < 81:
        parsed_numbers.insert(0, 0)
    return parsed_numbers

quiz = data_to_sudoku(data['quizzes'][0])
solution = data_to_sudoku(data['solutions'][0])
quiz_grid = gridding_array(quiz)
solution_grid = gridding_array(solution)

print("before solving:")
display_grid(quiz_grid)
print("\nafter solving:")
quiz_grid = backtrack(quiz_grid)
display_grid(quiz_grid)
if compare(quiz_grid, solution_grid):
  print("\nthe result is correct")
else:
  print("No solution found for this sudoku.")
```

```python
from ortools.sat.python import cp_model
import pandas as pd
import numpy as np
```

```python
def solve_sudoku(sudoku):
    """
    TODO: Solves the given Sudoku puzzle using ortools and returns the solved grid.

    :param sudoku_grid: 2D numpy array representing the Sudoku puzzle
    :return: Solved 2D numpy array or a message indicating failure
    """
    model = cp_model.CpModel()
    cells = [[model.NewIntVar(1, 9, f'cell_{i}_{j}') for j in range(9)] for i in range(9)]
    for i in range(9):
        model.AddAllDifferent(cells[i])
        model.AddAllDifferent([cells[j][i] for j in range(9)])
    for box_row in range(3):
        for box_col in range(3):
            box = [cells[box_row * 3 + i][box_col * 3 + j] for i in range(3) for j in range(3)]
            model.AddAllDifferent(box)
    for i in range(9):
        for j in range(9):
            if sudoku[i][j] != 0:
                model.Add(cells[i][j] == sudoku[i][j])
    solver = cp_model.CpSolver()
    status = solver.Solve(model)
    if status in (cp_model.FEASIBLE, cp_model.OPTIMAL):
        solution = np.array([[solver.Value(cells[i][j]) for j in range(9)] for i in range(9)])
        return solution
    print("No solution found for this sudoku.")
    return None
```

### AC-3 Algorithm (20 Points)
First and foremost, implement the AC-3 algorithm to be later used in the problem

```python
def AC3(csp, neighbors):
    """
    csp:
        dict{
            nodes: list of nodes,
            domains: dictionary {variable: [domain]},
            constraints: dictionary {(node_i, node_j): list of constraints
        },
    neighbors:
        dict {var: [neighbor]}

    TODO: Implement the AC3 algorithm

    return: a bool to show whether the problem is consistent and if so, the reduced domain of the problem
    """

    queue = [(xi, xj) for xi in csp['nodes'] for xj in neighbors[xi]]
    steps = 0

    while queue:
        (xi, xj) = queue.pop(0)
        steps += 1
        if revise(csp, xi, xj):
            for xk in neighbors[xi]:
                if xk != xj:
                    queue.append((xk, xi))
    return True, csp['domains'], steps

def revise(csp, xi, xj):
    revised = False
    domain_xi = csp['domains'][xi]
    domain_xj = csp['domains'][xj]

    for x in domain_xi:
        if not any((x, y) in csp['constraints'][(xi, xj)] for y in domain_xj):
            domain_xi.remove(x)
            revised = True
    return revised
```

### AC-4 Algorithm (20 Points)
Secondly, implement the AC-4 (Arc Consistency 4) algorithm. This algorithm enforces arc consistency by maintaining **support tables** and efficiently processing **a queue of arcs** for consistency checks. To help you implement it, here are the key steps and concepts to focus on:
1. **Support Tables**: For each pair of variable-value assignments, you must maintain a table that tracks the number of supporting values in the neighboring variables' domains. This table will allow you to quickly determine if a value is consistent with its neighbors
2. Start by initializing your support tables for all variable-value pairs. For each arc (X, Y), compute and store the number of values in Y's domain that support each value in X's domain based on the constraint between X and Y.
3. Initialize **a queue of arcs** with all pairs of variables that share a constraint
4. For each arc (X, Y) in the queue, check the support for each value in X's domain. If any value in X loses all supporting values in Y, remove that value from X’s domain.
5. If a value is removed from X’s domain, for every neighboring variable Z (other than Y), you must add the arc (Z, X) back into the queue for further consistency checks. This ensures that removing a value from X doesn’t create inconsistencies for other variables.
6. Whenever a value is removed from a domain, update the support tables for all neighboring variable-value pairs that depend on the removed value. This way, if a value of Z is only supported by the value that was removed from X, the support count for that value of Z should be decremented.
7. The algorithm terminates when the queue is empty, meaning all arcs have been processed and the domains of all variables are arc-consistent. If at any point a domain becomes empty, the problem is unsolvable.

```python
def AC4(csp):
    """
    AC-4 algorithm to enforce arc consistency using support tables.

    Parameters:
    - csp: A dictionary with nodes, domains, and constraints.

    Returns:
    - (bool, dict): A tuple where the first value indicates if the CSP is arc-consistent, and the second is the reduced domain.
    """
    support = {}
    queue = []
    steps = 22

    for (xi, xj), constraint in csp['constraints'].items():
        support[(xi, xj)] = {x: 0 for x in csp['domains'][xi]}
        for x in csp['domains'][xi]:
            support[(xi, xj)][x] = sum(1 for y in csp['domains'][xj] if (x, y) in constraint)
            steps += 1

    for xi in csp['nodes']:
        for xj in neighbors[xi]:
            for x in csp['domains'][xi]:
                steps += 1
                if support[(xi, xj)][x] == 0:
                    queue.append((xi, x))

    while queue:
        xi, x = queue.pop(0)
        steps += 1
        if x in csp['domains'][xi]:
            csp['domains'][xi].remove(x)
            if not csp['domains'][xi]:
                return False, None, steps

            for xk in neighbors[xi]:
                for z in csp['domains'][xk]:
                    if (z, x) in csp['constraints'][(xk, xi)]:
                        support[(xk, xi)][z] -= 1
                        steps += 1
                        if support[(xk, xi)][z] == 0:
                            queue.append((xk, z))
                            steps += 1

    return True, csp['domains'], steps
```

## Backtracking + AC-3 and AC-4 algorithm (20 Points)
Now we are going to utilize the implemented algorithms to speed up the backtracking process, use both of the above algorithms, you must run ac3 and ac4 algorithms seperately.

```python
def is_consistent(var, value, assignment, csp, neighbors):
    """
    Check if assigning `value` to `var` is consistent with current assignment.
    """
    for neighbor in neighbors[var]:
        if neighbor in assignment and not (value, assignment[neighbor]) in csp["constraints"][(var, neighbor)]:
            return False
    return True

def select_unassigned_variable(assignment, domains):
    """
    Select the unassigned variable with the smallest domain (MRV heuristic).
    """
    unassigned_vars = [var for var in domains if var not in assignment]
    return min(unassigned_vars, key=lambda var: len(domains[var]))

def backtracking_search(csp, neighbors, ac3):
    """
    Backtracking search with AC-3 or AC-4 pruning to solve the CSP.
    """
    def backtrack(assignment, ac3):
        if ac3:
          reduced_domains = AC3(csp, neighbors)[1]
        else :
          reduced_domains = AC4(csp)[1]
        if len(assignment) == len(csp["nodes"]):
            return assignment
        var = select_unassigned_variable(assignment, reduced_domains)
        for value in reduced_domains[var]:
            if is_consistent(var, value, assignment, csp, neighbors):
                assignment[var] = value
                result = backtrack(assignment, ac3)
                if result != None:
                    return result
                assignment.pop(var)
        return None
    return backtrack({}, ac3)

csp = {
    "nodes": nodes,
    "domains": domains,
    "constraints": constraints
}

# TODO: Solve using AC3 algorithm

print("\nBacktracking + AC3 algorithm: \n")
start = time()
solution = backtracking_search(csp, graph, True)
s = time() - start

if solution:
    print("Solution found:", solution)
else:
    print("No solution found!")

print("Solution retrieved in: %.4g seconds" % s)


# TODO: Solve using AC4 algorithm

print("\nBacktracking + AC4 algorithm:\n")
start = time()
solution = backtracking_search(csp, graph, False)
s = time() - start

if solution:
    print("Solution found:", solution)
else:
    print("No solution found!")

print("Solution retrieved in: %.4g seconds" % s)
```
